## **Project Summary: Exploratory Data Analysis of FedEx Shipment Dataset**

The goal of this project was to perform a comprehensive exploratory data analysis (EDA) on a shipment dataset provided by FedEx. This dataset contains detailed information related to the delivery and logistics process, including order details, vendor and client interactions, shipment modes, and financial attributes. The purpose of the analysis was to uncover meaningful insights, trends, and patterns in the data that could help stakeholders enhance business operations, reduce delays, and improve efficiency.

### **1. Data Understanding**

The dataset contains 10,324 rows and 53 columns, covering various aspects such as shipment scheduling, dates related to procurement and delivery, vendors, product types, dosage forms, financial metrics like line item value and freight cost, and country-wise distribution. Initial analysis revealed missing values in columns like `Shipment Mode`, `Dosage`, and `Line Item Insurance (USD)`. The dataset also contained redundant or inconsistent values that needed to be cleaned for meaningful interpretation.

### **2. Data Wrangling and Preprocessing**

The first step involved cleaning column names by converting them to lowercase and removing spaces or special characters for consistency. Five key date columns were identified and converted to `datetime` format, followed by the extraction of temporal features such as year, month, day, and weekday from each date field. Missing values were handled by appropriate imputation strategies: the mode was used for categorical columns like `Shipment Mode`, while the median and placeholder values were applied for numerical and text-based fields like `Line Item Insurance` and `Dosage`.

Duplicate rows were also removed, and certain categorical columns such as `Shipment Mode`, `Managed By`, and `Vendor INCO Term` were converted to the `category` type for better memory efficiency and easier analysis. The final cleaned dataset maintained the same number of columns but was significantly better structured and analysis-ready.

### **3. Data Visualization & Insights**

A series of charts were created to better understand the data. Bar plots showed the frequency of shipments by country and shipment mode, revealing that countries like Vietnam and Côte d'Ivoire had the highest shipment volume. A pie chart indicated a major preference for "Air" shipping mode, which could imply higher urgency or limited ground infrastructure.

A line plot of delivery counts over time highlighted seasonal trends in deliveries, while a box plot of freight costs exposed the presence of outliers and suggested the need to optimize high-cost shipments. The correlation heatmap provided a macro-level view of how numerical features like `Line Item Value`, `Pack Price`, and `Freight Cost` relate to one another. Lastly, a pair plot visualized inter-feature distributions, aiding in identifying patterns or potential feature engineering opportunities for predictive modeling.

These insights help businesses understand what drives costs and delays. For example, if specific countries are associated with higher freight costs or late deliveries, FedEx can re-evaluate vendor relationships or explore alternative shipping methods for those regions.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(" Column names standardized.\n")

# Function to identify date columns (more flexible for FedEx structure)
def is_date_column(col):
    return 'date' in col

# Identify date columns
date_columns = [col for col in df.columns if is_date_column(col)]
print(" Found date columns:", date_columns)

# Process each date column
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

    # Extract temporal features
    df[f'{col}_year'] = df[col].dt.year
    df[f'{col}_month'] = df[col].dt.month
    df[f'{col}_day'] = df[col].dt.day
    df[f'{col}_weekday'] = df[col].dt.weekday

print("\n Date columns converted and temporal features extracted.\n")

# Remove duplicates
df = df.drop_duplicates()
print(f" Removed duplicates. Current dataset shape: {df.shape}\n")

# Handling missing values (custom logic for FedEx dataset)
df['shipment_mode'].fillna(df['shipment_mode'].mode()[0], inplace=True)
df['dosage'].fillna('Unknown', inplace=True)
df['line_item_insurance_(usd)'].fillna(df['line_item_insurance_(usd)'].median(), inplace=True)

print(" Missing values handled for key columns.\n")

# Convert appropriate columns to 'category' dtype
categorical_columns = [
    'country', 'managed_by', 'fulfill_via', 'vendor_inco_term',
    'shipment_mode', 'product_group', 'sub_classification',
    'vendor', 'brand', 'dosage_form', 'unit_of_measure_(per_pack)',
    'manufacturing_site', 'first_line_designation'
]

for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype('category')

print(" Converted selected columns to categorical type.\n")

# Final shape and preview
print(" Final dataset shape:", df.shape)
df.head()
````

* Column names were cleaned and standardized.

* All 5 date columns were correctly identified and converted to datetime.

* Temporal features like year, month, day, and weekday were successfully extracted.

* Duplicates were removed, leaving 10,324 rows.

* Missing values were handled in key columns.

* Specified columns were successfully converted to categorical types.
